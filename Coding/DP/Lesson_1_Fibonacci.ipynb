```python
def recursive_step(n):
    if n < 0:
        raise ValueError("Input must be a non-negative integer")
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return recursive_step(n - 1) + recursive_step(n - 2)
    
recursive_step(100)
```

```python
def recursive_step_with_memorization(n,cache={}):
    if n < 0:
        raise ValueError("Input must be a non-negative integer")
    if n == 0:
        return 1
    elif n == 1:
        return 1
    elif n in cache:
        return cache[n]
    else:
        return_value = recursive_step_with_memorization(n - 1) + recursive_step_with_memorization(n - 2)
        cache[n] = return_value
        return return_value

recursive_step_with_memorization(100)
```

```python
def recursive_step_with_tabular(n):
    dp = [0] * (n+1)
    dp[0] = 1
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]

recursive_step_with_tabular(100)
    
```

```python
def recursive_step_with_tabular_optimise(n):
    a = 1
    b = 1
    for i in range(2, n+1):
        res = a + b
        a = b
        b = res
    return res


recursive_step_with_tabular_optimise(100)
```
